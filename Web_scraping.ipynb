### Hands-On

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Send a GET request to the web page to be scraped
url = "https://zh.wikipedia.org/zh-tw/臺北醫學大學"
response = requests.get(url)
print(response.status_code)
```

```python
# Create a BeautifulSoup object from the HTML content of the web page
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
elements = soup.find_all("span", {"class": "mw-headline"})

for i in elements:
    print(i.text)
```

```python
# Import required modules
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
#import chromedriver_autoinstaller
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

#Install chrome driver automatically
#chromedriver_autoinstaller.install()

# Create ChromeOptions object and add desired arguments
chrome_options = Options()
#chrome_options.add_argument("--headless") # run Chrome in headless mode (without GUI)
#chrome_options.add_argument("--disable-gpu") # disable GPU acceleration
#chrome_options.add_argument("--disable-notifications") #disable pop up windows

driver = webdriver.Chrome('chromedriver', options = chrome_options)

driver.get("https://zh.wikipedia.org/zh-tw/臺北醫學大學")

elements = driver.find_elements(By.CLASS_NAME, "mw-headline")


for element in elements:
    print(element.text)

time.sleep(2)
driver.close()
```

## Add switch menu

```python
chrome_options = Options()
# chrome_options.add_argument("--headless")

driver = webdriver.Chrome('chromedriver', options = chrome_options)
driver.set_window_size(1920, 1080)
driver.get("https://www.5music.com.tw/")

search_bar = driver.find_element(By.ID, "stxt")
search_bar.click()
search_bar.send_keys("le sserafim")
search_bar.send_keys(Keys.ENTER)

all_titles = []
all_dess = []


while True:
    titles = driver.find_elements(By.CLASS_NAME, "h4")
    title_db = [title.text for title in titles]
    # for title in titles:
    #     print(title.text)
    all_titles.extend(title_db)

    dess = driver.find_elements(By.CLASS_NAME, "des")
    des_db = [des.text for des in dess]
    all_dess.extend(des_db)
    
    switch_menu = driver.find_elements(By.CLASS_NAME, "next") 
    if len(switch_menu) >= 2:
        switch_menu[0].click()
    else:
        break

    
# i = 0
# for k in range(len(titles)):
#     print(titles[k].text+":"+dess[k].text)
#     i+=1

time.sleep(4)
driver.close()
```

```python
import pandas as pd

db_dict = {"Artist": all_titles, "Album name": all_dess}

df = pd.DataFrame.from_dict(db_dict)
print(df)
```
